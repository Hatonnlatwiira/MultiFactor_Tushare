```python
import tushare as ts
import pandas as pd
import datetime
ts.set_token('930681ab6294f8b1fe8a19e09b96e37e0b5923c955d179d522b79e37')
pro = ts.pro_api()
```

```python
start_dt = '20190101'
time_temp = datetime.datetime.now() - datetime.timedelta(days=1)
end_dt = time_temp.strftime('%Y%m%d')
#end_dt = '20191031'
print('Time period: from ' + start_dt + ' to ' + end_dt + '.')
```

```python
# 交易日历
trade_cal_SSE = pro.trade_cal(exchange='SSE', start_date=start_dt, end_date=end_dt, is_open=1)
trade_cal_SZSE = pro.trade_cal(exchange='SZSE', start_date=start_dt, end_date=end_dt, is_open=1)
trade_period = list(trade_cal_SSE.cal_date)
trade_period
```

```python
# 股票信息
# 退市股票被忽略了
stock_basic = pro.stock_basic(exchange='', list_status='L', fields='ts_code,symbol,name,area,industry,list_status,list_date,delist_date')
stock_pool = list(stock_basic.ts_code)
stock_num = len(stock_pool)
stock_basic.head()
```

```python
daily_basic = pd.DataFrame()
for i in trade_period:
    try:
        daily_basic = daily_basic.append(pro.daily_basic(ts_code='', trade_date=i, fields='ts_code,trade_date,turnover_rate,\
        turnover_rate_f,volume_ratio,pe,pe_ttm,pb,ps,ps_ttm,dv_ratio,dv_ttm,total_mv,circ_mv'))
        print('Date: ' + i)
    except:
        print('Date: ' + i + ' Data Error')
daily_basic.sort_values(['ts_code', 'trade_date'], inplace=True)
daily_basic.shape
```

```python
fina_indicator = pd.DataFrame()
pro_bar = pd.DataFrame()
for i in range(stock_num):
    try:
        fina_indicator = fina_indicator.append(pro.fina_indicator_vip(ts_code=stock_pool[i], start_date=start_dt, end_date=end_dt))
        pro_bar = pro_bar.append(ts.pro_bar(ts_code=stock_pool[i], adj='qfq', start_date=start_dt, end_date=end_dt))
        print('Seq: ' + str(i+1) + ' of ' + str(stock_num) + '   Code: ' + str(stock_pool[i]))
    except:
        print('Seq: ' + str(i+1) + ' of ' + str(stock_num) + '   Code: ' + str(stock_pool[i]) + ' Data Error')
fina_indicator.head()
```

```python
fina_indicator_col = ['ts_code', 'ann_date', 'end_date', 'profit_dedt', 'roe', 'roa']
fina_indicator = fina_indicator.loc[:,fina_indicator_col]
fina_indicator
# 财务数据拼接日度数据
```

```python
# 根据代码、日期排好序再输出
daily_basic.to_csv('daily_basic.csv', index=0)
fina_indicator.to_csv('fina_indicator.csv', index=0)
pro_bar.to_csv('pro_bar.csv', index=0)
```
